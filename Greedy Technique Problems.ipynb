# QUESTION: Minimum Absolute Difference in an Array

```python
'''

Min. Absolute Difference In Array

Given an integer array A of size N, find and return the minimum absolute difference between any two elements in the array.
We define the absolute difference between two elements ai, and aj (where i != j ) is |ai - aj|.
Input format :
Line 1 : Integer N, Array Size
Line 2 : Array elements (separated by space)
Output Format :
Minimum difference
Constraints :
1 <= N <= 10^6
Sample Input :
5
2 9 0 4 5
Sample Input :
1

'''
```

```python
def min_abs_diff(n,arr):
    arr=sorted(arr)
    mini=arr[1]-arr[0]
    for i in range(len(arr)-1):
        diff=abs(arr[i+1]-arr[i])
        #print(diff)
        if(diff<mini):
            mini=diff
        else:
            continue
    return mini
n=int(input())
holder=list(int(i) for i in input().strip().split(' '))
print(min_abs_diff(n,holder))

        
```

# QUESTION: Nikunj and Donuts

```python
'''

Nikunj and Donuts

Nikunj loves donuts, but he also likes to stay fit. He eats n donuts in one sitting, and each donut has a calorie count, ci. After eating a donut with k calories, he must walk at least 2^j x k(where j is the number donuts he has already eaten) miles to maintain his weight.
Given the individual calorie counts for each of the n donuts, find and print a long integer denoting the minimum number of miles Nikunj must walk to maintain his weight. Note that he can eat the donuts in any order.
Input
The first line contains an integer, n, denoting the number of donuts. 
The second line contains n space-separated integers describing the respective calorie counts of each donut I, i.e ci.
Output
Print a long integer denoting the minimum number of miles Nikunj must walk to maintain his weight.
Constraints
1 ≤ n ≤ 40
1 ≤ ci ≤ 1000
Sample Input
3
1 3 2
Sample Output
11

'''
```

```python
def nikunj(n,arr):
    arr=sorted(arr)
    sumi=0
    for i in range(len(arr)):
        sumi+=arr[i]*(2**(n-(i+1)))
    return sumi
n=int(input())
holder=list(int(i) for i in input().strip().split(' '))
print(nikunj(n,holder))
```

# Information about degeneracy:

If a, b, and c are the lengths of the three sides of a triangle, then

a+b>c

a+c>b

b+c>a

If any one of these inequalities is not true, then we get a degenerate triangle.

In simple term,first sort the a,b,c in ascending order then check below condition

a+b<=c

if this condition satisfy then triangle is degenerate triangle.

```python
def non_degenerate_triangle(sticks):
    sticks.sort()
    usable=[]
    for i in range(len(sticks)-2):
        if(sticks[i]+sticks[i+1]<=sticks[i+2]):
            continue
        else:
            usable.append((sticks[i],sticks[i+1],sticks[i+2]))
    if(len(usable)>0):
        return usable
    else:return -1
n=int(input())
sticks=list(int(i) for i in input().strip().split(' '))
final_sides=non_degenerate_triangle(sticks)
sums=[]
if(final_sides!=-1):
    for i in range(len(final_sides)):
        sumi=final_sides[i][0]+final_sides[i][1]+final_sides[i][2]
        sums.append(sumi)
        sumi=0
    target=max(sums)
    for i in range(len(sums)):
        if(sums[i]==target):
            reqd=i
            break
    print(final_sides[reqd][0],end=' ')
    print(final_sides[reqd][1],end=' ')
    print(final_sides[reqd][2],end=' ')
else:
    print(-1)
```

```python
def knapSack(W, wt, val, n): 
  
    # Base Case 
    if n == 0 or W == 0 : 
        return 0
  
    # If weight of the nth item is more than Knapsack of capacity 
    # W, then this item cannot be included in the optimal solution 
    if (wt[n-1] > W): 
        return knapSack(W, wt, val, n-1) 
  
    # return the maximum of two cases: 
    # (1) nth item included 
    # (2) not included 
    else: 
        return max(val[n-1] + knapSack(W-wt[n-1], wt, val, n-1), 
                   knapSack(W, wt, val, n-1)) 
  
# end of function knapSack 
  
# To test above function 
val = [1,1,1] 
wt = [1,6,9] 
W = 12
n = len(val) 
print (knapSack(W, wt, val, n)) 
```

# QUESTION: Problem Discussion

```python
'''

Problem discussion
Send Feedback
Harshit gave Aahad an array of size N and asked to minimize the difference between the maximum value and minimum value by modifying the array under the condition that each array element either increase or decrease by k(only once).
It seems difficult for Aahad so he asked for your help
Input Format
The First line contains two space-separated integers: N,K
Next lines contain N space-separated integers denoting elements of the array
Output Format
The output contains a single integer denoting the minimum difference between maximum value and the minimum value in the array
Constraints
1 =< N <= 10^5 
1 <= Ai,K <= 10^9
Sample Input1:
3 6
1 15 10
Sample Output1:
5
Explaination
We change from 1 to 6, 15 to  9 and 10 to 4. Maximum difference is 5 (between 4 and 9). We can't get a lower difference.

'''
```

```python
def problem_discussion(arr,k):
    arr.sort()
    big=arr[len(arr)-1]-k
    small=arr[0]+k
    if(small>big):
        temp=big
        big=small
        small=temp
    for i in range(1,len(arr)-1):
        subt=arr[i]-k
        add=arr[i]+k
        if(subt>=small or add<=big):
            continue
        elif(big-subt<=add-small):
            small=subt
            add=big
    return min((big-small),(arr[len(arr)-1]-arr[0]))
holder=list(int(i) for i in input().strip().split(' '))
arr=list(int(i) for i in input().strip().split(' '))
print(problem_discussion(arr,holder[1]))
```
